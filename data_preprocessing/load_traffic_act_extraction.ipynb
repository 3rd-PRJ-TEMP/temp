```python
from pypdf import PdfReader

# PDF 파일 경로 설정
pdf_path = '../data/Standards_of_Negligence_Ratio.pdf'

lines_with_keyword = []

reader = PdfReader(pdf_path)
for page in reader.pages:
    text = page.extract_text()

    if text:
        lines = text.split('\n')

        for line in lines:
            if '도로교통법 제' in line:
                lines_with_keyword.append(line)

print(len(lines_with_keyword))
for keyword in lines_with_keyword:
    print(keyword)
```

```python
import re

pattern = r'도로교통법 제 \d+조 | 도로교통법 제\d+조 | 도로교통법 제\d+조의\d+ | 도로교통법 제\d+조의 \d+'

raw_list = []

for line in lines_with_keyword:
    found = re.findall(pattern, line)
    raw_list.extend(found)

raw_list = list(set(raw_list))

for raw in sorted(raw_list):
    print(raw)
```
